# Importação dos dados:

`````python
import pandas as pd
import matplotlib.pyplot as plt
`````

`````python
loja1 = pd.read_csv("loja1.txt")
loja2 = pd.read_csv("loja2.txt")
loja3 = pd.read_csv("loja3.txt")
loja4 = pd.read_csv("loja4.txt")

loja1.head()
`````

# 1. Análise do faturamento:

`````python
faturamento1 = loja1["Preço"].sum()
faturamento2 = loja2["Preço"].sum()
faturamento3 = loja3["Preço"].sum()
faturamento4 = loja4["Preço"].sum()
`````

`````python
print(f" Faturamento da loja 1: {faturamento1} reais\n",
      f"Faturamento da loja 2: {faturamento2} reais\n", 
      f"Faturamento da loja 3: {faturamento3} reais\n", 
      f"Faturamento da loja 4: {faturamento4} reais\n")
`````

# 2. Vendas por Categoria

`````python
categoria1 = loja1.groupby("Categoria do Produto")[["Produto"]].count().sort_values('Produto', ascending=False)
categoria2 = loja2.groupby("Categoria do Produto")[["Produto"]].count().sort_values('Produto', ascending=False)
categoria3 = loja3.groupby("Categoria do Produto")[["Produto"]].count().sort_values('Produto', ascending=False)
categoria4 = loja4.groupby("Categoria do Produto")[["Produto"]].count().sort_values('Produto', ascending=False)
`````

`````python
print(categoria1.head())
print(categoria2.head())
print(categoria3.head())
print(categoria4.head())
`````

# 3. Média de Avaliação das Lojas

`````python
avaliacao1 = loja1["Avaliação da compra"].mean()
avaliacao2 = loja2["Avaliação da compra"].mean()
avaliacao3 = loja3["Avaliação da compra"].mean()
avaliacao4 = loja4["Avaliação da compra"].mean()
`````

`````python
print(f" Média das avaliações da loja 1: {avaliacao1}\n",
      f"Média das avaliações da loja 2: {avaliacao2}\n", 
      f"Média das avaliações da loja 3: {avaliacao3}\n", 
      f"Média das avaliações da loja 4: {avaliacao4}\n")
`````

# 4. Produtos Mais e Menos Vendidos

`````python
produtos1 = loja1.groupby("Produto")[["Categoria do Produto"]].count().sort_values('Categoria do Produto', ascending=False)
produtos2 = loja2.groupby("Produto")[["Categoria do Produto"]].count().sort_values('Categoria do Produto', ascending=False)
produtos3 = loja3.groupby("Produto")[["Categoria do Produto"]].count().sort_values('Categoria do Produto', ascending=False)
produtos4 = loja4.groupby("Produto")[["Categoria do Produto"]].count().sort_values('Categoria do Produto', ascending=False)

produtos1 = pd.concat([produtos1.head(3), produtos1.tail(3)])
produtos2 = pd.concat([produtos2.head(3), produtos2.tail(3)])
produtos3 = pd.concat([produtos3.head(3), produtos3.tail(3)])
produtos4 = pd.concat([produtos4.head(3), produtos4.tail(3)])
`````

`````python
print(produtos1)
print(produtos2)
print(produtos3)
print(produtos4)
`````

# 5. Frete Médio por Loja

`````python
frete1 = loja1["Frete"].mean()
frete2 = loja2["Frete"].mean()
frete3 = loja3["Frete"].mean()
frete4 = loja4["Frete"].mean()
`````

`````python
print(f" Média do frete da loja 1: {frete1}\n",
      f"Média do frete da loja 2: {frete2}\n", 
      f"Média do frete da loja 3: {frete3}\n", 
      f"Média do frete da loja 4: {frete4}\n")
`````

# Gráficos:

`````python
# Dados
faturamentos = [faturamento1, faturamento2, faturamento3, faturamento4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Gráfico
plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos)
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
`````

`````python
fig, axs = plt.subplots(2, 2, figsize=(12,8))
categorias = [categoria1, categoria2, categoria3, categoria4]
titulos = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

for i, ax in enumerate(axs.flat):
    categorias[i].plot(kind='bar', ax=ax, legend=False)
    ax.set_title(f'Vendas por Categoria - {titulos[i]}')
    ax.set_xlabel('Categoria do Produto')
    ax.set_ylabel('Quantidade Vendida')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
`````

`````python
avaliacoes = [avaliacao1, avaliacao2, avaliacao3, avaliacao4]

plt.figure(figsize=(8,5))
plt.bar(lojas, avaliacoes, color='orange')
plt.title('Média de Avaliação das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Nota Média')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
`````

`````python
fig, axs = plt.subplots(2, 2, figsize=(12,8))
produtos = [produtos1, produtos2, produtos3, produtos4]

for i, ax in enumerate(axs.flat):
    produtos[i].plot(kind='bar', ax=ax, legend=False)
    ax.set_title(f'Produtos Mais e Menos Vendidos - Loja {i+1}')
    ax.set_xlabel('Produto')
    ax.set_ylabel('Quantidade Vendida')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
`````

`````python
fretes = [frete1, frete2, frete3, frete4]

plt.figure(figsize=(8,5))
plt.bar(lojas, fretes, color='green')
plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Valor Médio do Frete (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
`````
